```python
def prepare_frame(df):
    """
    Подготавливает dataset для дальнейшей работы 
    """
    # преобразовываю названия колонок
    clean_columns_names = [re.sub('\s','',column_name) for column_name in df.columns]
    # переименовываю имена колонок и удаляю столбец с датой (по условию он не нужен)
    clients_df = df.rename(
                            columns = dict(zip(df.columns,clean_columns_names))
                        ).drop('Date', axis=1)
    return clients_df
```

```python
def search_params(fc, sc):
    """
    Производит поиск id по которым были 'пересечения'
    """
    # группирую по колонке которая является первым принмаемым параметром
    df = clients_df[[fc,sc]].groupby([fc]).agg(set).reset_index()
    check_list, res_list = [], []
    #прохожу по всем элементам сгруппированных данных
    for set_ in df[sc].tolist():
        for elem in set_:
            # если элемент уже есть в списке для проверки - значит он дублируется и нам подходит
            if elem in check_list and not elem in res_list:
                res_list.append(elem)
            else:
                check_list.append(elem)
    return {sc : res_list}
```

```python
def create_pars(search_list, key):
    """
        Функция для группировки данных по ключу и преобразования в строку для конкатенации 
        с запросом к базе данных
    """
    res_list = []
    for element in search_list:
        try:
            res_list.extend(element[key])
        except:
            pass
    #преобразование set в строку с помощью регулярных выражений
    prep = re.sub('{','(',str(set(res_list)))
    prep = re.sub('}',')',prep)
    return prep
```

```python
def create_pars(search_list, key,  variant):
    """
        Функция для группировки данных по ключу и преобразования в строку для конкатенации 
        с запросом к базе данных
    """
    res_list = []
    for element in search_list:
        try:
            res_list.extend(element[key])
        except:
            pass
    if variant == 1:
        return list((set(res_list)))
    elif variant== 2:
        prep = re.sub('{','(',str(set(res_list)))
        prep = re.sub('}',')',prep)
        return prep
    else:
        print("Выбран неверный вариант")
```

```python
#формируются все возможные комбинации пар стобцов для поиска 'пересечений'
vals = ['fingerprintId', 'cid', 'user_id', 'ClientID']
perm_set = itertools.permutations(vals,2)
new_list = []
for elem in perm_set:
    new_list.append(search_params(elem[0], elem[1]))
```

```python
def clean_data(string):
    regexes = [
        r'\&.*', # удалет часть запроса к браузеру
        r"\(.*\)", #все что заключено в кавычки
        r"\d+\s*с.зон",
        r"\d+\s*серия",
        r"\d+\s*выпуск",
        r"\d+\s*глав.",
        r"\d+\s*часть",
        r"\d{1,2}\s+[а-я]{3,}(а|я)\s+\d+\s+год",# день месяц yyyy года
        r"эфир\s*от\s*\d+.{1,2}\d+.{1,2}\d+", # эфир от dd.mm.yyyy
        r"\d+.{1,2}\d+.{1,2}\d+",# все даты
        r"https:\S+",
        r"\d{1,2}[:\s{1}]\d{1,2}", # время в формате hh:mm или hh mm
        r"^:\/\/.*",# удаляет данные в формате://rutube.dev/video/871397215457216624b35345eb466b69/
        r"[^a-zA-Zа-яА-Я0-9]",
        r"\s+"  
    ]
    replace_gram = {
        'solovievlive': 'соловьев лайв',
        r'с.ло.ьев' : 'соловьев',
        'solovyov': 'соловьев',
        'solovjev' : 'соловьев',
        'solovjov' : 'соловьев',
        'soloviev' : 'соловьев',
        'соловьв' : 'соловьев',
        'soloviov':'соловьев',
        'live' : 'лайв',
        'life': 'лайв',
        'лайф' : 'лайв',
        r'онл.йн' : 'онлайн',
        'сенодня' : 'сегодня',
        'андрец' : 'андрей',
        'шоппинг': 'шопинг',
        'новостм': 'новости',
        'соловьев лав':'соловьев лайв'
    }
    for regex in regexes:
        string = re.sub(regex,' ', string)
    for key in replace_gram.keys():
        string = re.sub(key, replace_gram[key], string)
    return string
```

```python
stop_words = stopwords.words('russian')
morph = pymorphy2.MorphAnalyzer()

def prepare_text(line):
    tl = nltk.word_tokenize(line)
    nft = [morph.parse(token)[0].normal_form for token in tl if len(token)>1]#nft - normal form token
    clean_words_list = [normalToken for normalToken in nft if normalToken not in stop_words and len(normalToken)>1]#преобразованный список
    return ' '.join(clean_words_list)
```
